### Regressing out confounds and recomputing correlation

#### Counfounds:
 * Target Size
 * SNR (RSM diagonal mean)
 * Intervertex distance (average across vertices in parcels)

```python
region_size = utils.get_region_size()
meansub_RSM_SNR = utils.get_RSM_SNR()
intervertex_dist_LR_parcelmean = utils.get_parcelwise_intervertex_distance()
```

```python
# Regressing out confounds from meansub_SVPR

X = np.hstack([region_size.reshape(-1,1),
               meansub_RSM_SNR.reshape(-1,1),
               intervertex_dist_LR_parcelmean.reshape(-1,1)])
y = meansub_SVPR.reshape(-1,1)

coef,resid = utils.get_residual(X,y)
```

```python
x = np.squeeze(resid)
y = meansub_dim_diff

obs_r, pval = utils.get_brainsmash_correlation(x,y,n_perm=1000)
print('obs_r = ',obs_r, ', pval = ',pval)
```
