## Figure 3C: Dimensionality prediction

```python
obs_dim = utils.get_obs_dimensionality()
meansub_obs_dim = np.mean(obs_dim,axis=0)

pred_dim = utils.get_pred_dimensionality()
meansub_pred_dim = np.mean(pred_dim,axis=0)

meansub_dim_error_idx = np.abs(meansub_pred_dim - meansub_obs_dim)
```

```python
pred_dim_connperm = utils.get_pred_dimensionality_connperm()
obs_dim_tiled = np.tile(obs_dim[:, :, np.newaxis], (1, 1, 100))
dim_error_idx_connperm = np.abs(pred_dim_connperm - obs_dim_tiled)

perm_data = dim_error_idx_connperm
meansub_dim_error_idx_groupnull = utils.generate_group_null_from_perm(perm_data, n_groupnull = 1000, nPerm = 100)
```

```python
# Sorted scatterplot with null

observed_values = meansub_dim_error_idx/meansub_obs_dim
null_distributions = meansub_dim_error_idx_groupnull/meansub_obs_dim.reshape(-1,1) # to ensure element-side division in tiled case
pvalues = utils.get_pval_parcelwise(observed_values,null_distributions,direction='one-sided-lower',n_null=1000)

file_out = 'dim_error_idx'
ylabel = 'Dim. error index'

fig = plot_utils.plot_sorted_scatterplot_with_null(observed_values, null_distributions, pvalues, ylabel, file_out, title=None, alpha=0.05, 
                                         figsize=(10,5), region_subset=None, sort=True)
```

```python
# Brain plot with significant values only

sig_idx = np.where(pvalues < 0.05)
observed_values_sig = np.where(pvalues < 0.05, observed_values, -9999)
print('No. of significant regions (out of 360):',np.sum(pvalues < 0.05))
```

```python
inputdata = observed_values_sig
file_out = 'dim_error_idx'
title = 'Dimensionality Prediction Error'

plot_utils.get_brain_plot(inputdata,file_out,title,ignore_unvalid=True)
```

## Figure 3D: Compression-then-expansion in double-CV RSMs

```python
doubleCV_dim = utils.get_doubleCV_dimensionality()
meansub_doubleCV_dim = np.mean(doubleCV_dim,axis=0)
```

```python
doubleCV_dim_surr = utils.get_doubleCV_dimensionality_surr(n_surrogates = 100)

# Generating group null
perm_data = doubleCV_dim_surr
n_groupnull = 1000
doubleCV_dim_surr_groupnull = utils.generate_group_null_from_perm(perm_data, n_groupnull = n_groupnull, nPerm = 100)
```

```python
# BS_avg: groupnull

nBrainSys = 3

doubleCV_dim_surr_groupnull_BSavg = np.zeros((nBrainSys,n_groupnull))
for groupnullIdx in range(n_groupnull):
    
    data = doubleCV_dim_surr_groupnull[:,groupnullIdx]
    doubleCV_dim_surr_groupnull_BSavg[:,groupnullIdx] = utils.get_BS_averaged_data(data)
```

```python
# BS_avg: observed
data = meansub_doubleCV_dim
meansub_doubleCV_dim_BSavg = utils.get_BS_averaged_data(data)
```

```python
# Sensory - Association

sen_asso_obs_diff = meansub_doubleCV_dim_BSavg[0] - meansub_doubleCV_dim_BSavg[1]

sen_asso_null_diff = np.zeros(n_groupnull)
for groupnullIdx in range(n_groupnull):
    sen_asso_null_diff[groupnullIdx] = (doubleCV_dim_surr_groupnull_BSavg[0,groupnullIdx] - 
                                        doubleCV_dim_surr_groupnull_BSavg[1,groupnullIdx])
        
obs_value = sen_asso_obs_diff
null_hist_data = sen_asso_null_diff
xlabel = 'Double CV Dim.'
title = 'Sensory_minus_Asso_doubleCV_dim_BSavg_hist'

fig = plot_utils.plot_null_hist(obs_value, null_hist_data, xlabel, title)

# p-value:
direction = 'one-sided-higher'
pval = utils.get_pval(obs_value,null_hist_data,direction,n_null=1000)
print('p-value:',pval)
```

```python
# Association - Motor

asso_motor_obs_diff = meansub_doubleCV_dim_BSavg[1] - meansub_doubleCV_dim_BSavg[2]

asso_motor_null_diff = np.zeros(n_groupnull)
for groupnullIdx in range(n_groupnull):
    asso_motor_null_diff[groupnullIdx] = (doubleCV_dim_surr_groupnull_BSavg[1,groupnullIdx] - 
                                        doubleCV_dim_surr_groupnull_BSavg[2,groupnullIdx])
        
obs_value = asso_motor_obs_diff
null_hist_data = asso_motor_null_diff
xlabel = 'Double CV Dim.'
title = 'Asso_minus_Motor_doubleCV_dim_BSavg_hist'

fig = plot_utils.plot_null_hist(obs_value, null_hist_data, xlabel, title)

# p-value:
direction = 'one-sided-lower'
pval = utils.get_pval(obs_value,null_hist_data,direction,n_null=1000)
print('p-value:',pval)
```
