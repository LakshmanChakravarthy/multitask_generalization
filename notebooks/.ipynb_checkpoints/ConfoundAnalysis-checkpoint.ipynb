# Post-PNAS confound analyses

### Counfounds:
 * Target Size
 * SNR (RSM diagonal mean)
 * Intervertex distance (average across vertices in parcels)

```python
import pickle
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.linear_model import LinearRegression
# from sklearn import metrics as skm
import statsmodels.api as sm

import wbplot
import matplotlib.image as img
import matplotlib.colors as clrs
from matplotlib import colorbar

# import RSMtools
# import plottools

from tqdm import tqdm

from brainsmash.mapgen.base import Base
from brainsmash.mapgen.stats import pearsonr, pairwise_r
from brainsmash.mapgen.stats import nonparp
```

```python
projdir = '/home/ln275/f_mc1689_1/multitask_generalization/'
subProjDir = projdir + 'data/derivatives/RSM_ActFlow/'
surfaceDeriv_dir = projdir + 'data/derivatives/surface/'
helpfiles_dir = projdir + 'docs/experimentfiles/'
figoutdir = projdir + 'docs/figures/working/vertexwiseFC/'
fcdir = projdir + 'data/derivatives/FC_new/'
```

```python
subIDs=['02','03','06','08','10','12','14','18','20',
        '22','24','25','26','27','28','29','30','31']

onlyRestSubIdx = [ 0, 1, 3, 4, 6, 7, 8,10,11,
                  12,20,13,21,14,22,15,23,16] # Use when dealing with full subjects' data (n=24)

nSub = len(subIDs)
nParcels = 360
nVertices = 59412
nSessions = 2
nTaskCond = 96

glasserfilename = helpfiles_dir + 'Q1-Q6_RelatedParcellation210.LR.CorticalAreas_dil_Colors.32k_fs_RL.dlabel.nii'

glasser = np.squeeze(nib.load(glasserfilename).get_fdata())

networkdef = np.loadtxt(helpfiles_dir + 'cortex_parcel_network_assignments.txt')
networkNames = ['VIS1','VIS2','SMN','CON','DAN','LAN','FPN','AUD','DMN','PMULTI','VMM','ORA']
```

### Load data

```python
# Region sizes
region_size = np.zeros(nParcels)
for roiIdx in range(nParcels):
    region_size[roiIdx] = len(np.where(glasser==roiIdx+1)[0])

# Parcel-level FC
with open(fcdir + 'allSubFC_parCorr.pkl', 'rb') as f:
    allSubParCorr = pickle.load(f)
    
# Representational dimensionality differences
with open(subProjDir + 'allsub_dimensionality_activeVisual_observed_18sub.pkl', 'rb') as f:
    allsub_obs_dimensionality = pickle.load(f)
print('allsub_obs_dimensionality shape:', allsub_obs_dimensionality.shape)

dim_diff = np.zeros((nSub,nParcels))
mean_connreg_dim = np.zeros((nSub,nParcels))

for subIdx in range(nSub):    
    for targetroi_idx in range(nParcels):

        # get connected regions indices
        connected_regionsIdx = np.where(allSubParCorr[subIdx,targetroi_idx,:]!=0)[0]
        
        # get mean dimensionality of connected regions
        mean_connreg_dim[subIdx,targetroi_idx] = np.mean(allsub_obs_dimensionality[subIdx,connected_regionsIdx])
        dim_diff[subIdx,targetroi_idx] = allsub_obs_dimensionality[subIdx,targetroi_idx] - mean_connreg_dim[subIdx,targetroi_idx]
print('dim_diff (target-meansource) shape:', dim_diff.shape)
meansub_dim_diff = np.mean(dim_diff,axis=0)
    
# Computing RSM SNR:
with open(subProjDir + 'allsub_taskCond_RSM_activeVisual.pkl', 'rb') as f:
    allsub_taskCondRSM = pickle.load(f)
    
select_sub_taskCondRSM = allsub_taskCondRSM[onlyRestSubIdx,:,:,:] # rest data's available only for 18 sub
print('select_sub_taskCondRSM shape:', select_sub_taskCondRSM.shape)

# Computing mean of RSM diagonal as SNR metric:
allsub_RSM_SNR = np.zeros((nSub,nParcels))

for subIdx in range(nSub):
    for regIdx in range(nParcels):
        allsub_RSM_SNR[subIdx,regIdx] = np.trace(select_sub_taskCondRSM[subIdx,regIdx,:,:])/nTaskCond
print('allsub_RSM_SNR shape:',allsub_RSM_SNR.shape)
meansub_RSM_SNR = np.mean(allsub_RSM_SNR,axis=0)

# FC dimensionality
with open(subProjDir + 'allsub_alltargetSVPR.pkl', 'rb') as f:
    allsub_SVPR = pickle.load(f)
print('allsub_SVPR shape:', allsub_SVPR.shape)
meansub_SVPR = np.mean(allsub_SVPR,axis=0)

# Intervertex distance
with open(surfaceDeriv_dir + 'intervertex_dist_LR.pkl', 'rb') as f:
    intervertex_dist_LR = pickle.load(f)
    
intervertex_dist_LR_parcelmean = np.zeros(nParcels)
for roiIdx in range(nParcels):
    intervertex_dist_LR_parcelmean[roiIdx] = np.nanmean(intervertex_dist_LR[np.where(glasser==roiIdx+1)[0]])
print('intervertex_dist_LR_parcelmean shape:',intervertex_dist_LR_parcelmean.shape)
```

```python
file_out = figoutdir + "intervertex_dist_LR_parcelmean.png"
inputdata = intervertex_dist_LR_parcelmean

#flip hemispheres, since CAB-NP is ordered left-to-right, while wbplot uses right-to-left
inputdata_flipped=np.zeros(np.shape(inputdata))
inputdata_flipped[0:180]=inputdata[180:360]
inputdata_flipped[180:360]=inputdata[0:180]

#Set to all reds if no negative values
# if min(inputdata) >= 0:
#     colormap='Reds'
# else:
#     colormap='seismic'

vmin, vmax = np.min(inputdata), np.max(inputdata)

colormap='Reds'
wbplot.pscalar(
        file_out=file_out,
        pscalars=inputdata_flipped,
        vrange = (vmin,vmax),
        cmap=colormap,
        transparent=True)

plt.figure(figsize=(3.5,3))
ax = plt.subplot(111)
im = img.imread(file_out) 
plt.imshow(im)
plt.axis('off')
plt.title('Intervertex distance',fontsize=18)

# vmin, vmax = -2, 2
cnorm = clrs.Normalize(vmin=vmin, vmax=vmax)  # only important for tick placing
cmap = plt.get_cmap('Reds')
cax = ax.inset_axes([0.44, 0.48, 0.12, 0.07])
cbar = colorbar.ColorbarBase(
    cax, cmap=cmap, norm=cnorm, orientation='horizontal')
#cbar.set_ticks([-2, 2])  # don't need to do this since we're going to hide them
cax.get_xaxis().set_tick_params(length=0, pad=-2)
cbar.set_ticklabels([])
cbar.outline.set_visible(False)
cax.text(-0.025, 0.4, str(round(vmin,2)), ha='right', va='center', transform=cax.transAxes,
         fontsize=12);
cax.text(1.025, 0.4, str(round(vmax,2)), ha='left', va='center', transform=cax.transAxes,
         fontsize=12);
plt.tight_layout()
plt.savefig(figoutdir + 'intervertex_dist_LR_parcelmean.pdf',transparent=True) 
```

```python
# Direct correlation
# Correlation with confounds

X = meansub_SVPR.reshape(-1,1)
y = meansub_dim_diff.reshape(-1,1)
    
model = LinearRegression()
model.fit(stats.zscore(X,axis=0), stats.zscore(y,axis=0))

print(model.coef_)
```

```python
# Regression with confounds together

X = np.hstack([meansub_SVPR.reshape(-1,1),
               region_size.reshape(-1,1),
               meansub_RSM_SNR.reshape(-1,1),
               intervertex_dist_LR_parcelmean.reshape(-1,1)])
y = meansub_dim_diff.reshape(-1,1)
    
model = LinearRegression()
model.fit(stats.zscore(X,axis=0), stats.zscore(y,axis=0))

observed_coef = np.array(model.coef_[0])

print(observed_coef)
```

```python
X1 = np.hstack([region_size.reshape(-1,1),
               meansub_RSM_SNR.reshape(-1,1),
               intervertex_dist_LR_parcelmean.reshape(-1,1)])
X2 = meansub_SVPR.reshape(-1,1)
    
model = LinearRegression()
model.fit(X1, X2)
coef = np.array(model.coef_)

X2_pred = np.dot(coef,X1.T)
resid = X2 = X2_pred.T

y = meansub_dim_diff.reshape(-1,1)

model2 = LinearRegression()
model2.fit(stats.zscore(resid,axis=0), stats.zscore(y,axis=0))

print(model2.coef_)
```

```python
# Stats on FC dim coef, with region size as covariate

allsub_dimdiff_vs_FCdim_covariate = np.zeros((nSub,4))

for subIdx in range(nSub):

    X = np.hstack([allsub_SVPR[subIdx,:].reshape(-1,1),
                   region_size.reshape(-1,1),
                   allsub_RSM_SNR[subIdx,:].reshape(-1,1),
                   intervertex_dist_LR_parcelmean.reshape(-1,1)])
    y = dim_diff[subIdx,:].reshape(-1,1)

    model = LinearRegression()
    model.fit(stats.zscore(X,axis=0), stats.zscore(y,axis=0))

    allsub_dimdiff_vs_FCdim_covariate[subIdx,:] = model.coef_[0]
    
print('Mean coef. across subjects:',np.mean(allsub_dimdiff_vs_FCdim_covariate,axis=0))

# t-test against zero:
t,p = stats.ttest_1samp(allsub_dimdiff_vs_FCdim_covariate, popmean=0, axis=0)

print('t-values:',t,'\n','p-values:',p)
```

```python
# Brainsmash statistical testing

# All brain regions

np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning)
n_surrogates = 1000
geo_dist_rh = np.loadtxt(helpfiles_dir + 'RightParcelGeodesicDistmat.txt')
geo_dist_lh = np.loadtxt(helpfiles_dir + 'LeftParcelGeodesicDistmat.txt')
base = Base(meansub_dim_diff[:180],geo_dist_lh)
surrogates_lh = base(n=n_surrogates)
base = Base(meansub_dim_diff[180:],geo_dist_rh)
surrogates_rh = base(n=n_surrogates)
surrogates_dim = np.hstack((surrogates_lh,surrogates_rh))
```

```python
surrogates_dim.shape
```

```python
# Regression on surrogates

X1 = np.hstack([region_size.reshape(-1,1),
               meansub_RSM_SNR.reshape(-1,1),
               intervertex_dist_LR_parcelmean.reshape(-1,1)])
X2 = meansub_SVPR.reshape(-1,1)
    
model = LinearRegression()
model.fit(X1, X2)
coef = np.array(model.coef_)

X2_pred = np.dot(coef,X1.T)
resid = X2 = X2_pred.T

null_coef = []
for surrogate_idx in range(n_surrogates):

    y = surrogates_dim[surrogate_idx,:].reshape(-1,1)

    model2 = LinearRegression()
    model2.fit(stats.zscore(resid,axis=0), stats.zscore(y,axis=0))

    null_coef.append(model2.coef_[0])
```

```python
null_coef = np.array(null_coef)
null_coef.shape
```

```python
fig, axes = plt.subplots(figsize=(5, 4))

sns.histplot(null_coef, kde=True, color='skyblue', ax=axes)
axes.axvline(x=0.315, color='red', linestyle='--', linewidth=2)
axes.set_title(f'Rep. Exansion/Compression vs. FC Dim (Pearson r)')
axes.set_xlabel('Coefficient Value')
axes.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
null_distributions = null_coef

# Observed regression coefficients
observed_values = 0.315

regressor labels = ['']

# Set up the figure with 4 subplots (2x2 grid)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()  # Flatten to easily iterate

# Create histograms for each regressor
for i in range(4):
    # Plot histogram of null distribution using seaborn
    sns.histplot(null_distributions[:, i], kde=True, color='skyblue', ax=axes[i])
    
    # Add vertical line for observed value
    axes[i].axvline(x=observed_values[i], color='red', linestyle='--', linewidth=2)
    
    # Add label for the vertical line
    axes[i].text(
        observed_values[i], 
        axes[i].get_ylim()[1]*0.9, 
        f'Observed: {observed_values[i]:.2f}', 
        color='red',
        horizontalalignment='right' if observed_values[i] < 0 else 'left',
        verticalalignment='top',
        rotation=90,
        fontweight='bold'
    )
    
    # Add title and labels
    axes[i].set_title(f'Regressor {i+1}')
    axes[i].set_xlabel('Coefficient Value')
    axes[i].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```
