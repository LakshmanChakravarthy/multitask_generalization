## Degree of compression-to-expansion: sensorimotor-to-association ratio

```python
def get_relative_compression_metric(dims):
    # dims: 360 element vector of dims for 360 regions
    sensory_roi_id,association_roi_id,motor_roi_id,_,_ = RSMtools.getBSwise_ROIid()
    relative_compression_metric = np.mean([np.mean(dims[sensory_roi_id]),np.mean(dims[motor_roi_id])]) / np.mean(dims[association_roi_id])
    
    return relative_compression_metric
```

```python
X = all_parcelwise_dims_BSavg
y = all_scores

def leave_one_out_cv_regression(X, y):
    n_samples = len(y)
    predictions = np.zeros(n_samples)
    # For storing betas from each iteration
    # +1 for the constant term
    all_betas = np.zeros((n_samples, X.shape[1] + 1))
    
    X_with_const = sm.add_constant(X)
    
    for i in range(n_samples):
        mask = np.ones(n_samples, dtype=bool)
        mask[i] = False
        
        X_train = X_with_const[mask]
        y_train = y[mask]
        X_val = X_with_const[~mask]
        
        model = sm.OLS(y_train, X_train).fit()
        predictions[i] = model.predict(X_val)[0]
        
        # Store coefficients from this iteration
        all_betas[i, :] = model.params
    
    # Calculate mean and std of betas across iterations
    mean_betas = np.mean(all_betas, axis=0)
    std_betas = np.std(all_betas, axis=0)
    
    correlation, p_value = stats.pearsonr(y, predictions)
    
    return predictions, correlation, p_value, all_betas, mean_betas, std_betas

# Usage:

predictions, correlation, p_value, all_betas, mean_betas, std_betas = leave_one_out_cv_regression(X, y)
```

### Multiple regression: attempting cross-validation across subs despite underconstrained nature

```python
def perform_pca_once(X):
    """
    Standardize data and perform PCA once, return all components
    """
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    pca = PCA()
    
    # ## Get nPCs that give me 80% variance explained ##
    # pca.fit(X_scaled)
    # # Get explained variance ratios
    # explained_variance_ratio = pca.explained_variance_ratio_
    # cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
    # # Find number of components needed
    # n_components = np.argmax(cumulative_variance_ratio >= 0.8) + 1
    
    pc_scores = pca.fit_transform(X_scaled)
    return pc_scores, pca, scaler

def inner_cv_for_pc_selection(pc_scores, y, n_folds=9):
    """
    Perform 9-fold CV to find optimal number of PCs using progressive addition
    """
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
    max_pcs = 50 #instead of pc_scores.shape[1]
    r_squared_per_pc = np.zeros((n_folds, max_pcs))

    for fold_idx, (train_idx, val_idx) in enumerate(kf.split(pc_scores)):
        X_train, X_val = pc_scores[train_idx], pc_scores[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]

        # Try different numbers of PCs progressively
        for n_pc in range(1, max_pcs + 1):
            model = sm.OLS(y_train, sm.add_constant(X_train[:, :n_pc])).fit()
            y_pred = model.predict(sm.add_constant(X_val[:, :n_pc]))
            r_squared = r2_score(y_val, y_pred)
            r_squared_per_pc[fold_idx, n_pc-1] = r_squared

    mean_r_squared = np.mean(r_squared_per_pc, axis=0)
    optimal_n_pcs = np.argmax(mean_r_squared) + 1

    return optimal_n_pcs, mean_r_squared

def transform_betas_to_original_space(pca, betas, n_components, n_features):
    """
    Transform PC space betas back to original feature space
    
    Parameters:
    -----------
    pca : sklearn.decomposition.PCA
        Fitted PCA object
    betas : numpy.ndarray
        Beta coefficients from PC space regression (excluding intercept)
    n_components : int
        Number of components used in the regression
    n_features : int
        Number of features in original space
        
    Returns:
    --------
    numpy.ndarray
        Beta coefficients in original feature space
    """
    # Get the transformation matrix (eigenvectors)
    components = pca.components_[:n_components]
    
    # Transform betas back to original space
    # Note: we multiply by components.T because PCA.components_ gives eigenvectors as rows
    original_betas = np.zeros(n_features)
    original_betas = np.dot(components.T, betas)
    
    return original_betas

def leave_one_out_cv_with_betas(X, y):
    """
    Leave-one-out CV with optimal PC selection using inner 9-fold CV
    Returns predictions and betas in original feature space
    
    Parameters:
    -----------
    X : numpy.ndarray
        Feature matrix
    y : numpy.ndarray
        Target variable
        
    Returns:
    --------
    dict
        Dictionary containing predictions, correlation, p-value,
        optimal number of PCs, and average betas in original space
    """
    n_samples, n_features = X.shape
    predictions = np.zeros(n_samples)
    optimal_pcs_used = np.zeros(n_samples)
    
    # Store betas for each iteration
    all_betas = np.zeros((n_samples, n_features))
    all_intercepts = np.zeros(n_samples)
    
    for i in range(n_samples):
        
        print(i)
        
        # Split into train and test
        train_idx = list(set(range(n_samples)) - {i})
        X_train, X_test = X[train_idx], X[i:i+1]
        y_train, y_test = y[train_idx], y[i]
        
        # Perform PCA on training data
        pc_scores, pca, scaler = perform_pca_once(X_train)
        
        # Find optimal number of PCs using inner CV
        optimal_n_pcs, _ = inner_cv_for_pc_selection(pc_scores, y_train)
        optimal_pcs_used[i] = optimal_n_pcs
        
        # Fit final model using optimal number of PCs
        final_pc_scores = pc_scores[:, :optimal_n_pcs]
        final_model = sm.OLS(y_train, sm.add_constant(final_pc_scores)).fit()
        
        # Store the intercept
        all_intercepts[i] = final_model.params[0]
        
        # Transform betas back to original space
        pc_betas = final_model.params[1:]  # Exclude intercept
        original_betas = transform_betas_to_original_space(
            pca, pc_betas, optimal_n_pcs, n_features
        )
        all_betas[i] = original_betas
        
        # Transform test data and predict
        X_test_scaled = scaler.transform(X_test)
        X_test_pc = pca.transform(X_test_scaled)[:, :optimal_n_pcs]
        X_test_pc_wConst = np.insert(X_test_pc, 0, 1, axis=1)
        
        predictions[i] = final_model.predict(X_test_pc_wConst)[0]
    
    # Calculate final correlation
    correlation, p_value = stats.pearsonr(y, predictions)
    
    # Calculate average betas and their standard deviations
    avg_betas = np.mean(all_betas, axis=0)
    std_betas = np.std(all_betas, axis=0)
    avg_intercept = np.mean(all_intercepts)
    
    return {
        'predictions': predictions,
        'correlation': correlation,
        'p_value': p_value,
        'optimal_pcs': optimal_pcs_used,
        'average_betas': avg_betas,
        'std_betas': std_betas,
        'average_intercept': avg_intercept,
        'all_betas': all_betas,  # Including all iterations for further analysis
        'all_intercepts': all_intercepts
    }

# Usage:

X = all_parcelwise_dims
y = all_scores
results = leave_one_out_cv_with_betas(X, y)
```
